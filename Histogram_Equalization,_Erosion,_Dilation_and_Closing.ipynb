```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Histogram Equalization

```python
# reading image in Gray
image = cv2.imread('/content/trees.jpg', 0)
```

```python
print(image.shape)
```

```python
plt.imshow(image)
```

```python
result = cv2.equalizeHist(image)
plt.imshow(result)
```

Erosion

```python
j = cv2.imread('/content/j.jpg', 0)
kernel = np.ones((3,3), np.uint8)
plt.imshow(j)
```

```python
eroded_image = cv2.erode(j, kernel, iterations=1)
plt.imshow(eroded_image)
```

Dilation

```python
j = cv2.imread('/content/j.jpg', 0)
kernel = np.ones((3,3), np.uint8)
plt.imshow(j)
```

```python
eroded_image = cv2.dilate(j, kernel, iterations=1)
plt.imshow(eroded_image)
```

Opening

```python
j = cv2.imread('/content/j.jpg', 0)
kernel = np.ones((3,3), np.uint8)
plt.imshow(j)
```

```python
opening = cv2.morphologyEx(j, cv2.MORPH_OPEN, kernel, iterations=1)
plt.imshow(opening)
```

Closing

```python
j = cv2.imread('/content/j.jpg', 0)
kernel = np.ones((3,3), np.uint8)
plt.imshow(j)
```

```python
closing = cv2.morphologyEx(j, cv2.MORPH_CLOSE, kernel, iterations=1)
plt.imshow(closing)
```
